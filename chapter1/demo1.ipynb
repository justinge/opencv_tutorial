## BGR顺序

```python
import cv2
img_logo = cv2.imread("../images/logo.png")
## 注意, imgpath 入参如果不是真实的图片,就会返回None
# NoneType type(img_logo) 
```

```python
img_logo.shape
```

```python
b,g,r = cv2.split(img_logo)
b,g,r # 获取整张图片的b,g,r
```

```python
# 手动方式转成RGB
img_new = cv2.merge([r, g, b])
```

```python
# 显示图片
cv2.imshow('img_logo',img_logo)
cv2.imshow('img_new1',img_new)

# 等待
cv2.waitKey(0)
# 关闭所有窗口
cv2.destroyAllWindows()
```

## plt 的show 是RGB,所以其实输入之前要做cv2的2RGB的转换

```python
import matplotlib.pyplot as plt
plt.subplot(121)
plt.imshow(img_logo)
plt.subplot(122)
plt.imshow(img_new)
plt.show()
```

```python
img_logo_new = cv2.cvtColor(img_logo, cv2.COLOR_BGR2RGB) # BGR -> RGB
```

```python
import matplotlib.pyplot as plt
plt.subplot(121)
plt.imshow(img_logo)
plt.subplot(122)
plt.imshow(img_logo_new)
plt.show()
```

## 两种方式来挑整通道顺序
img_logo_new = cv2.cvtColor(img_logo, cv2.COLOR_BGR2RGB) # BGR -> RGB

## 手动方式转成RGB
b,g,r = cv2.split(img_logo)
img_new = cv2.merge([r, g, b])
