```python
#1.导入库
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2.设置字体颜色
colors = {
    'blue':(255,0,0),
    'green':(0,255,0),
    'red':(0,0,255),
    'yellow':(0,255,255),
    'gray':(125,125,125),
    'dark_gray':(50,50,50),
    'light_gray':(220,220,220),
    'white':(255,255,255),
    'cyan':(255,255,0)
}
```

```python
# 3.方法 显示图片 
def show_image(image, title):
    # BGR to RGB
    img_RGB = image[:,:,::-1]
    plt.title(title)
    plt.imshow(img_RGB)
```

```python
# 4 创建画布
canvas = np.zeros((400, 400, 3),np.uint8) # 默认背景时黑色
canvas.fill(255) # 背景变白色
show_image(canvas, "Canvas")
```

```python
# 5. 往画布输入文本
cv2.putText(canvas,"Hello World", (50,50),cv2.FONT_HERSHEY_SIMPLEX,0.9,color=colors["blue"])
cv2.putText(canvas,"Welcome", (50,150),cv2.FONT_HERSHEY_SIMPLEX,1.4,color=colors["cyan"])
cv2.putText(canvas,"Justin", (50,250),cv2.FONT_HERSHEY_SIMPLEX,2.0,color=colors["dark_gray"])
cv2.putText(canvas,"我是谁", (50,350),cv2.FONT_HERSHEY_SIMPLEX,2.0,color=colors["dark_gray"])
show_image(canvas, "Canvas")
```
